```
data = [("James", "Sales", 3000), \
    ("Michael", "Sales", 4600), \
    ("Robert", "Sales", 4100), \
    ("Maria", "Finance", 3000), \
    ("James", "Sales", 3000), \
    ("Scott", "Finance", 3300), \
    ("Jen", "Finance", 3900), \
    ("Jeff", "Marketing", 3000), \
    ("Kumar", "Marketing", 2000), \
    ("Saif", "Sales", 4100), \
        ("Rangayya", "Sales", 4202), \
        ("Pullamma", "Marketing", 4400), \
        ("Rangayya", "Sales", 4202), \
        ("Ratish", "Finance", 8100), \
        ("Saif", "Finance", 7100), \
        ("Pullamma", "Marketing", 4400), \
        ("Saif", "Marketing", 4900), \
        ("Ratish", "Finance", 8100), \
        ("Ragamma", "Sales", 4600), \
        ("Ragamma", "Sales", 4600), \
        ("Tirupal", "Finance", 6600), \
        ("Tirupal", "Marketing", 4600), \
        ("Tirupal", "Sales", 7600), \
        ("Tirupal", "Sales", None), \
        ("Ramayya", None, 7600), \
        (None, "Sales", 7600) \
  ]
columns= ["employee_name", "department", "salary"]
df = spark.createDataFrame(data = data, schema = columns)
df.printSchema()
df.show(30)
```

1. Dropping by using distinct()

```
df_dist = df.distinct()
print('distinct count: '+str(df_dist.count()))
df_dist.show()
```

2.
drop using dropDuplicates()

```
df_drpDup = df.dropDuplicates();
print('distinct count: '+str(df_drpDup.count()))
df_drpDup.show()
```

3.
drop using dropDuplicates(['column1','column2',...])

```
df_dDc = df.dropDuplicates(['employee_name'])
print('distinct count'+str(df_dDc.count()))
df_dDc.show()
```

```
df_dDc2 = df.dropDuplicates(['department'])
print('distinct count'+str(df_dDc.count()))
df_dDc.show()
```

4. 
creating a new dataframe for all null valued rows and unaccepted format.
here we remove the null occured rows, remove duplicates.

```
df_w = df.filter('employee_name is null OR department is null OR salary is null')
df_w.show()
```

```
df_r = df.na.drop(how='any', subset=['employee_name','department','salary']).dropDuplicates()
df_r.show()
```

5.
using a window function to eliminate duplicates and also record the duplicates

```
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number
windowconf = Window.partitionBy('employee_name','department').orderBy('salary')

df_w = df.withColumn('row_number', row_number().over(windowconf))
df_w.show(50,truncate=False)
```

```
df_wd = df_w.filter
df_wd.show()
```

```
df_wd = df_w.filter(df_w.row_number == 2).drop('row_number')
df_wd.show()
```

```
df_wr = df_w.filter(df_w.row_number == 1).drop('row_number')
df_wr.show()
```
